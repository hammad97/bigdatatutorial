* Matrix Multiplication

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
m = 20

matA = np.matrix(n * m)
matA = np.random.random((n, m))

vecV = np.arange(m* 1).reshape(m, 1)
vecV = np.random.normal(2, 0.01, size = (m, 1))
vecC = np.zeros((n, 1))

i = 0
while i < n:
    j = 0
    while j < m:
        vecC[i, 0] = vecC[i, 0] + (matA[i, j] * vecV[j, 0])
        j = j + 1
    i = i + 1

meanC = np.mean(vecC)
stdC = np.std(vecC)
print("Mean : " + str(meanC))
print("Standard Deviation : " + str(stdC))

plt.hist(vecC, bins = 5)
plt.title("Vector-C Histogram")
plt.show()
```

* Grading Problem

```python
import pandas as pd


grades_df = pd.read_csv('Grades.csv')
grades_df
```

```python
# I saw there were some negative numbers in the data and my assumption here is that it was some error 
#  in the dataset since marks cannot be negative so in this preprocessing step I made all the negative
#    numbers to zero.
grades_df.loc[grades_df["English"] < 0, "English"] = 0
grades_df.loc[grades_df[" Maths"] < 0, " Maths"] = 0
grades_df.loc[grades_df["Science"] < 0, "Science"] = 0
grades_df.loc[grades_df["German"] < 0, "German"] = 0
grades_df.loc[grades_df["Sports"] < 0, "Sports"] = 0

grades_df
```

```python
sub_cols = ['English', ' Maths', 'Science', 'German', 'Sports']
grades_df['Sum'] = grades_df[sub_cols].sum(axis = 1)
grades_df
```

```python
grades_df['Average'] = grades_df[sub_cols].mean(axis = 1)
grades_df
```

```python
grades_df['SD'] = grades_df[sub_cols].std(axis = 1)
grades_df
```

```python
grades_df.plot.bar(x = 'First Name', y = 'Average', rot = 0, figsize = (40,20))
```

```python
grdList =[]
for i in grades_df['Average']:
    if i >= 96 and i <= 100:
        grdList.append('A+')
    elif i >= 90 and i <= 95:
        grdList.append('A')
    elif i >= 86 and i <= 89:
        grdList.append('A-')
    elif i >= 80 and i <= 85:
        grdList.append('B+')
    elif i >= 76 and i <= 79:
        grdList.append('B')
    elif i >= 70 and i <= 75:
        grdList.append('B-')
    elif i >= 66 and i <= 69:
        grdList.append('C+')
    elif i >= 60 and i <= 65:
        grdList.append('C')
    elif i >= 56 and i <= 59:
        grdList.append('D')        
    else:
        grdList.append('F')
grades_df['Grade'] = grdList

numGrades = grades_df.groupby(['Grade']).size().reset_index(name = 'Number of Students')
numGrades
```

```python
plt.hist(numGrades['Grade'], weights = numGrades['Number of Students'], width = 0.75, bins = len(numGrades['Grade']))
```

```python
def learn_simple_linreg(x1, y1):
    b1 = np.mean((x1 - x1.mean()) * (y1 - y1.mean())) / np.mean((x1 - x1.mean()) ** 2)
    b0 = y1.mean() - b1 * x1.mean()
    return b0, b1

def pred_simple_linreg(x1, b0, b1):
    return x1 * b1 + b0

def plot_scatter(i, x1, y1, Pred_Y, sigma, b0, b1, axes):
    axes[i].plot(x1, Pred_Y, c = 'red', label = 'Prediction')
    axes[i].scatter(x1, y1, c = 'orange', label = 'Truth')
    axes[i].set_title('sigma = ' +  str(sigma) + '  (b0, b1) = (' + str(b0) + ', ' + str(b1) + ')')
    axes[i].set_xlabel('Values of X1')
    axes[i].set_ylabel('Values of Y1')
    axes[i].legend()
    return

sigmaList, betaList, sigma_ds = [0.01, 0.1, 1] , [], []

for sigma in sigmaList:
    sigma_ds.append(np.random.normal(scale = sigma, loc = 2, size = (100, 2)))

for i, (sigma, ds) in enumerate(zip(sigmaList, sigma_ds)):
    x1 = ds[:, 0]
    b0, b1 = learn_simple_linreg(x1, ds[:, 1])
    betaList.append(np.array([b0, b1]))

figure, axes = plt.subplots(3, 1, figsize = (8, 18))

for i, ((b0, b1), sigma, ds) in enumerate(zip(betaList, sigmaList, sigma_ds)):
    x1 = ds[:, 0]
    Pred_Y = pred_simple_linreg(x1, b0, b1)
    plot_scatter(i, x1, ds[:, 1], Pred_Y, sigma, b0, b1, axes)
    
figure, axes = plt.subplots(3, 1, figsize = (8, 18))

for i, ((b0, b1), sigma, ds) in enumerate(zip(betaList, sigmaList, sigma_ds)):
    x1 = ds[:, 0]
    b0 = 0
    Pred_Y = pred_simple_linreg(x1, b0, b1)
    plot_scatter(i, x1, ds[:, 1], Pred_Y, sigma, b0, b1, axes)

figure, axes = plt.subplots(3, 1, figsize = (8, 18))

for i, ((b0, b1), sigma, ds) in enumerate(zip(betaList, sigmaList, sigma_ds)):
    x1 = ds[:, 0]
    b1 = 0
    Pred_Y = pred_simple_linreg(x1, b0, b1)
    plot_scatter(i, x1, ds[:, 1], Pred_Y, sigma, b0, b1, axes)


linalgRes = []
for i, (sigma, ds) in enumerate(zip(sigmaList, sigma_ds)):
    x1 = ds[:, 0]
    one_stack = np.stack([np.ones_like(x1), x1], axis = 1)
    matA = one_stack.T @ one_stack
    matB = one_stack.T @ ds[:, 1]
    linalgRes.append(np.linalg.lstsq(matA, matB)[0])

figure, axes = plt.subplots(3, 1, figsize = (8, 18))

for i, ((b0, b1), sigma, ds) in enumerate(zip(linalgRes, sigmaList, sigma_ds)):
    x1 = ds[:, 0]
    Pred_Y = pred_simple_linreg(x1, b0, b1)
    plot_scatter(i, x1, ds[:, 1], Pred_Y, sigma, b0, b1, axes)
```
